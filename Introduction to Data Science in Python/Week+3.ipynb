# Merging Dataframes


首先我们来讨论一下如何合并Dataframes。还是以商店信息的DF为例：

```python
import pandas as pd

df = pd.DataFrame([{'Name': 'Chris', 'Item Purchased': 'Sponge', 'Cost': 22.50},
                   {'Name': 'Kevyn', 'Item Purchased': 'Kitty Litter', 'Cost': 2.50},
                   {'Name': 'Filip', 'Item Purchased': 'Spoon', 'Cost': 5.00}],
                  index=['Store 1', 'Store 1', 'Store 2'])
df
```

下面是常见的方法合并DF，直接使用`[]`

```python
df['Date'] = ['December 1', 'January 1', 'mid-May']
df
```

```python
df['Delivered'] = True
df
```

现在问题来了，如果我们并不清楚想要合并的数据，怎么办呢？当然，你可以直接使用`None` .

```python
df['Feedback'] = ['Positive', None, 'Negative']
df
```

下面介绍一种更加合理的方法：利用Series。

```python
adf = df.reset_index()
adf['Date'] = pd.Series({0: 'December 1', 2: 'mid-May'})
adf
```

上面方法的好处是：你只需要输入已知数值的行，其他行python自动填充为`None`.

以上，我们讨论了如何给DF添加一列数据，下面我们讨论一下如何合并两个DF，以学生信息和职工信息的DF为例，先看看数据：

```python
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])
staff_df = staff_df.set_index('Name')
student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally', 'School': 'Engineering'}])
student_df = student_df.set_index('Name')
print(staff_df.head())
print()
print(student_df.head())
```

合并两组DF使用`pd.merge`函数，主要需要确定两个参数，第一个参数是`how`，表示如何合并。第二个参数是`left_index`和`right_index`表示合并发生在那一列。
其中`how`参数主要有两种选项：`outer`和`inner`，分别表示合并采用‘和’的方式还是‘或’的方式。
举例说明：

```python
pd.merge(staff_df, student_df, how='outer', left_index=True, right_index=True)
```

首先，输入前两个参数分别是两个DF，注意这里第一个DF代表left，第二个代表right.
`how='outer'`代表采用和的方式，所以两个DF中出现的人名，在合并后的DF中都出现了。
`left_index=True, right_index=True`表示合并发生在那一列，这里的意思是发生在left的index也就是`name`那列，和right的index也是`name`那列。

```python
pd.merge(staff_df, student_df, how='inner', left_index=True, right_index=True)
```

上面改变了参数`how='inner'`，合并后的DF只有在两个DF都出现过的名字这里才出现。

当然，还可以选择`how='left'`或`how='right'`表示只保留left或right的DF的数据。

```python
pd.merge(staff_df, student_df, how='left', left_index=True, right_index=True)
```

```python
pd.merge(staff_df, student_df, how='right', left_index=True, right_index=True)
```

如果我们想要合并发生的列不是index怎么办？可以直接使用列的名字：

```python
staff_df = staff_df.reset_index()
student_df = student_df.reset_index()
pd.merge(staff_df, student_df, how='left', left_on='Name', right_on='Name')
```

如果两个DF中的数据发生冲突了怎么办？也就是说，如果student和staff中都包含列`Location`，合并后会发生什么？

```python
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR', 'Location': 'State Street'},
                         {'Name': 'Sally', 'Role': 'Course liasion', 'Location': 'Washington Avenue'},
                         {'Name': 'James', 'Role': 'Grader', 'Location': 'Washington Avenue'}])
student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business', 'Location': '1024 Billiard Avenue'},
                           {'Name': 'Mike', 'School': 'Law', 'Location': 'Fraternity House #22'},
                           {'Name': 'Sally', 'School': 'Engineering', 'Location': '512 Wilson Crescent'}])
pd.merge(staff_df, student_df, how='left', left_on='Name', right_on='Name')
```

如上，合并后增加了Location_x和Location_y，分别代表左边df和右边df的location。

除了对单个列进行合并，我们还可以让合并发生在多个列，如下，我们让合并发生在`'First Name'`和`'Last Name'`。

```python
staff_df = pd.DataFrame([{'First Name': 'Kelly', 'Last Name': 'Desjardins', 'Role': 'Director of HR'},
                         {'First Name': 'Sally', 'Last Name': 'Brooks', 'Role': 'Course liasion'},
                         {'First Name': 'James', 'Last Name': 'Wilde', 'Role': 'Grader'}])
student_df = pd.DataFrame([{'First Name': 'James', 'Last Name': 'Hammond', 'School': 'Business'},
                           {'First Name': 'Mike', 'Last Name': 'Smith', 'School': 'Law'},
                           {'First Name': 'Sally', 'Last Name': 'Brooks', 'School': 'Engineering'}])
staff_df
student_df
pd.merge(staff_df, student_df, how='inner', left_on=['First Name','Last Name'], right_on=['First Name','Last Name'])
```

结果是，只有'First Name'和'Last Name'在两个DF中都匹配了才出现在合并后的Df中，所以满足条件的只有‘Sally Brooks’。

# Idiomatic Pandas: Making Code Pandorable

开始写代码前，先讨论一下题目的意思：Idiomatic Pandas: Making Code Pandorable
当你想实现一个功能时，一般会有多种实现路径，最好的路径一般称之为idiomatic，具体来说idiomatic的方式，满足两点，即有很好的性能（high performance），还有很好的可读性(high readability).
那么，用panda实现一个功能的最佳路径被称为Pandorable。
所以题目的意思就是学会使用Pandas的最佳路径，Idiomatic Pandas: Making Code Pandorable。
下面进入举例说明：

先读入'census.csv'作为示例DF。

```python
import pandas as pd
df = pd.read_csv('census.csv')
df
```

接下来我们要实现的是：选出所有'SUMLEV'为50的行，删除有na值的行，将'STNAME'和'CTYNAME'设为index，将'ESTIMATESBASE2010'改名为'Estimates Base 2010'
以下使用两种实现路径，第一种是Pandorable，另一种是不好的方式。

```python
(df.where(df['SUMLEV']==50)
    .dropna()
    .set_index(['STNAME','CTYNAME'])
    .rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'}))
```

```python
df = df[df['SUMLEV']==50]
df.set_index(['STNAME','CTYNAME'], inplace=True)
df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})
```

我们发现，两种方式都能得到想要的结果，但是第一种方式更加Pandorable，具体来说，采用“链式调用”的方式比反复重复调用要好，而且不会占用更多的内存空间。   
这里还有一个技巧，用括号`()`把这个“链式调用”括起来，是为了让代码更加清晰和具有可读性。

下面介绍apply函数，apply函数能够将定义好的函数应用到DF中的每一行。   
举例来说，我们先定义函数`min_max()`，这个函数的作用是得到一行中我们关注的列的最大和最小值。

```python
import numpy as np
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min': np.min(data), 'max': np.max(data)})
```

现在我们将`min_max()`函数应用到df中的每一行   
注意，这里需要明确axis参数，这个参数为1代表对df中的行进行操作，0代表对列操作。

```python
df.apply(min_max, axis=1)
```

如果我们想要在得到最大最小值的同时，增加列来记录该值，可以修改`min_max`函数：

```python
import numpy as np
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['max'] = np.max(data)
    row['min'] = np.min(data)
    return row
df.apply(min_max, axis=1)
```

如何使用一行代码实现上面的方法？答案是利用lambada：

```python
rows = ['POPESTIMATE2010',
        'POPESTIMATE2011',
        'POPESTIMATE2012',
        'POPESTIMATE2013',
        'POPESTIMATE2014',
        'POPESTIMATE2015']
df.apply(lambda x: np.max(x[rows]), axis=1)
```

# Group by

接下来讨论`pd.groupby`，首先加载pandas和numpy库，然后读入census数据：

```python
import pandas as pd
import numpy as np
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
df
```

为什么说groupby是比循环更Pandorable的方法，我们用下面的例子来说明，首先我们采用循环的方式，求每个州人口的平均值。并记录时间。

```python

for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
```

```python

for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
```

groupby给DF分组的另一个好处是，可以用这种方式使处理器的不同进程分别处理分组的数据。   
例如在下面的例子里，我们首先将'STNAME'设为index，这样groupby默认对index进行分组，然后，我们定义一个函数fun(),如果传入的数据的第一个字母排在M前面，返回0，排在M和Q之间，返回1，否则返回2，这样，将fun函数输入groupby，就是按照每个州首字母将数据分为0，1，2三组。

```python
df.head()
```

```python
df = df.set_index('STNAME')

def fun(item):
    if item[0]<'M':
        return 0
    if item[0]<'Q':
        return 1
    return 2

for group, frame in df.groupby(fun):
    print('There are ' + str(len(frame)) + ' records in group ' + str(group) + ' for processing.')
```

最后讨论一下groupby的agg属性，agg就是aggregation的缩写，我们给agg传入一个词典，告诉他对那一列进行什么操作，将返回操作后的结果。   
具体来说，现在我们知道每个州2010年人口的平均值，首先利用groupby将数据按照州分组，然后利用agg传入{'CENSUS2010POP': np.average}，也就是对每组数据的'CENSUS2010POP'这一列求平均，最后得到每个州2010年人口的平均值。见下面代码。

```python
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
```

```python
df.groupby('STNAME').agg({'CENSUS2010POP': np.average})
```

这里讨论一下DataFrameGroupBy类型和SeriesGroupBy类型。   
groupby之后取出一列数据就是SeriesGroupBy类型，取出多列数据就是DataFrameGroupBy类型。

```python
print(type(df.groupby(level=0)['POPESTIMATE2010','POPESTIMATE2011']))
print(type(df.groupby(level=0)['POPESTIMATE2010']))
```

SeriesGroupBy类型如果求平均和求和的话，很简单，它会直接应用到这一列。

```python
(df.set_index('STNAME').groupby(level=0)['CENSUS2010POP']
    .agg({'avg': np.average, 'sum': np.sum}))
```

DataFrameGroupBy类型如果求和和求平均的话，会分别应用于选择的列，并返回一个‘分层’表示的DF。

```python
(df.set_index('STNAME').groupby(level=0)['POPESTIMATE2010','POPESTIMATE2011']
    .agg({'avg': np.average, 'sum': np.sum}))
```

注意，如果给agg传入的词典的关键字与列的名字重复了，那么python仅将函数应用于该列。

```python
(df.set_index('STNAME').groupby(level=0)['POPESTIMATE2010','POPESTIMATE2011']
    .agg({'POPESTIMATE2010': np.average, 'POPESTIMATE2011': np.sum}))
```

# Scales

接下来要讨论的概念是Scales数据类型，当我们处理数据的时候，数据类型一般可以定义为string、number等，但是这还不够。例如考试成绩评级，你有可能得到A+、A、B-等，如果将这种数据定义为string，那么我们将无法比较A+和B的大小，由于其不是数字，因此也无法将其定义为number。有没有一种数据类型可以定义类似的数据呢，答案是Scales。具体来说Scales有Ratio Scale、Initerval Scale、Ordinal Scale、Nominal Scale四种类型，本文先不深入讨论所有类型，只拿出一种常用的用法抛砖引玉。
下面我们定义一组DF，由成绩得分和对应的评价组成。

```python
df = pd.DataFrame(['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D'],
                  index=['excellent', 'excellent', 'excellent', 'good', 'good', 'good', 'ok', 'ok', 'ok', 'poor', 'poor'])
df.rename(columns={0: 'Grades'}, inplace=True)
df
```

利用`df.astype()`进行类型转换

```python
df['Grades'].astype('category').head()
```

我们看到`dtype: category`类型转换成功，共11个类别。如果我们增加参数`categories`和`ordered=True`,可以给数据排序。

```python
grades = df['Grades'].astype('category',
                             categories=['D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+'],
                             ordered=True)
grades.head()
```

如果我们想要成绩大于C的，可以非常简洁的使用比较运算符即可实现！

```python
grades > 'C'
```

这就是Scale类型相比其他类型方便之处。

生成Scale类型的数据有很多方法，这里介绍一种比较常见的方法：pd.cut()
通过这种方式可以将连续的数据切割成分段的数据：

```python
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
df = df.set_index('STNAME').groupby(level=0)['CENSUS2010POP'].agg({'avg': np.average})
pd.cut(df['avg'],10)
```

注意：读者在运行过程中可能会碰到如下警告：‘FutureWarning: using a dict on a Series for aggregation
is deprecated and will be removed in a future version’
这是因为panda版本0.2之前，采用传入词典的方式修改列的名称是可以的，但是之后的版本建议采用`.rename()`的方式修改名字。见下面的例子。

```python
# Create a sample data frame
df = pd.DataFrame({'A': [1, 1, 1, 2, 2],
                   'B': range(5),
                   'C': range(5)})

# ==== SINGLE COLUMN (SERIES) ====
# Syntax soon to be deprecated
df.groupby('A').B.agg({'foo': 'count'})
# Recommended replacement syntax
df.groupby('A').B.agg(['count']).rename(columns={'count': 'foo'})
```

# Pivot Tables

这一章节介绍Pivot_table，当你想了解一个DF不同列之间的变化关系时，这是个很好用的功能。我们读入‘cars.csv'用来举例：

```python
#http://open.canada.ca/data/en/dataset/98f1a129-f628-4ce4-b24d-6f16bf24dd64
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

观察上面的数据，如果我们想要知道每一年（YEAR）每个品牌(Make)的车辆的平均功率值(kW)，应该怎么实现呢？

答案是使用`.pivot_table()`，这里需要明确的参数包括`values`, `index`, `columns`, `aggfunc`，`values`代表我们关心的数据，也就是’kW‘这一列，`index`代表生成的DF的index,我们选择‘YEAR’这一列，`columns`选择汽车品牌‘Make’，`aggfunc`表示对不同年代不同品牌的`values`的操作，这里使用求平均。

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=np.mean)
```

我们看到，生成的表格清晰的展示了从2012年到2016年，每个汽车品牌的平均功率的变化。

另外，还有我们还可以给aggfunc参数传入不止一个函数，`.pivot_table()`会自动帮我们计算各个函数的结果，并用一个‘分层’的表格清晰的展示出来。

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=[np.mean,np.min], margins=True)
```

### Working with Dates in a Dataframe

利用`pd.data_range()`我们可以按照定义的方式生成‘DatetimeIndex’类型数据。
